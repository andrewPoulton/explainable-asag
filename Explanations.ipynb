```python
if False:
    for f in os.scandir('explained'):
        try:
            name = os.path.join('explained', f.name) 
            df = pd.read_pickle(os.path.join('explained', f.name))
            df['attr_score'] = df['attr_score'].apply(lambda x: x.item())
            df.to_pickle(os.path.join('explained', f.name))
        except:
            pass
    
```

```python
def get_data_record(df,i, aggr = 'norm'):
    row = df.iloc[i]
    r = visualization.VisualizationDataRecord(
                        row.get('attr_' + aggr, -1),
                        row.get('pred_prob', 0),
                        row.get('pred', -1),
                        row.get('label', -1),
                        row.get('attr_class', -1),
                        row.get('attr_score', 0),
                        row['tokens'].split(' '),
                        row.get('delta'))
    return r

def visualize(df, I, aggr = 'norm'):
    visualization.visualize_text([get_data_record(df,i, aggr = aggr) for i in I])
```

```python
df = pd.read_pickle(os.path.join('explained', 'bert-base_beetle_1fkyddwv_GradientShap.pkl'))
display(df.head())
visualize(df, range(10), aggr = 'norm')
```
