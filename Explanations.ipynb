```python
import pandas as pd
from captum.attr import (
    IntegratedGradients,
    visualization
    )
from explain import *
```

```python
def get_data_record(df,i):
    row = df.iloc[i]
    r = visualization.VisualizationDataRecord(
                        row.get('attr', -1),
                        row.get('pred_prob', 0),
                        row.get('pred', -1),
                        row.get('label', -1),
                        row.get('attr_class', -1),
                        row.get('attr_score', 0),
                        row['tokens'].split(' '),
                        row.get('delta'))
    return r
```

```python
df = pd.read_pickle('explained/roberta-base_IntegratedGradients.pkl')
visualization.visualize_text([get_data_record(df,i) for i in range(30,40)]);
```
